```python
# Problem 1

# Define functions that calculate the distance between two points within a square, circle, 
# cube and sphere
def dist_square(x1, x2, y1, y2):
    
    return(((x1-x2)**2 + (y1-y2)**2))**(1/2)

def dist_circle(x, y):
    
    return((x)**2 + (y)**2)

def dist_cube(x1, y1, z1, x2, y2, z2):
    
    return((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)**(1/2)

def dist_sphere(x, y, z):
    
    return((x)**2 + (y)**2 + (z)**2)

# define functions that give the result of each distance for a random sample number N
def results_square(a):
    d_sq = []
    for i in range(a):
        x1 = rand.uniform(0, 1) # random number generator b/w 0 and 1
        x2 = rand.uniform(0, 1)
        y1 = rand.uniform(0, 1)
        y2 = rand.uniform(0, 1)
        d_square = dist_square(x1, x2, y1, y2) # execute distance between 2 points given coordinates for 2 points
        d_sq.append(d_square)
    d_sq = np.array(d_sq)
    mean = sum(d_sq) / a
    var = sum((d_sq - mean)**2) / a
    unc = np.sqrt(var) / np.sqrt(a)
    return(mean, unc, var)


def results_circle(a):
    d_cir = []
    for i in range(a):
        x = rand.uniform(0, 1)
        y = rand.uniform(0, 1)
        d_circle = dist_circle(x, y)
        d_cir.append(d_circle)
    d_cir = np.array(d_cir)
    mean = sum(d_cir) / a
    var = sum((d_cir - mean)**2) / a
    unc = np.sqrt(var) / np.sqrt(a)
    return(mean, unc, var)
#def results_circle(a):
    #for i in range(a):
        #d_circle = []
        #yield dist_circle(rand.uniform(0, 1), rand.uniform(0, 1))
        

def results_cube(a):
    d_cbe = []
    for i in range(a):
        x1 = rand.uniform(0, 1)
        x2 = rand.uniform(0, 1)
        y1 = rand.uniform(0, 1)
        y2 = rand.uniform(0, 1)
        z1 = rand.uniform(0, 1)
        z2 = rand.uniform(0, 1)
        d_cube = dist_cube(x1, x2, y1, y2, z1, z2)
        d_cbe.append(d_cube)
    d_cbe = np.array(d_cbe)
    mean = sum(d_cbe) / a
    var = sum((d_cbe - mean)**2) / a
    unc = np.sqrt(var) / np.sqrt(a)
    return(mean, unc, var)
       
def results_sphere(a):
    d_sph = []
    for i in range(a):
        x = rand.uniform(0, 1)
        y = rand.uniform(0, 1)
        z = rand.uniform(0, 1)
        d_sphere = dist_sphere(x, y, z)
        d_sph.append(d_sphere)
    d_sph = np.array(d_sph)
    mean = sum(d_sph) / a
    var = sum((d_sph - mean)**2) / a
    unc = np.sqrt(var) / np.sqrt(a)
    return(mean, unc, var)

N = 1000000 # number of samples
print('Mean, uncertainty and variance for dist within square:', results_square(N))
print('Mean, uncertainty and variance for dist within circle:', results_circle(N))
print('Mean, uncertainty and variance for dist within cube:', results_cube(N))
print('Mean, uncertainty and variance for dist within sphere:', results_sphere(N))

```
